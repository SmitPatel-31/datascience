# Data Visulisation

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#line plotting
x=np.arange(10)
print (x)
y1=x**2+3
print(y1)
y2=2*x+3
print(y2)
```

```python
#to plot the graph
plt.plot(x,y1)
plt.show()
```

```python
plt.plot(x,y2)
plt.show()
```

```python
#to plot both in one 
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#to change the theme
```

```python
themes=plt.style.available
print(themes)
```

```python
# now assign any theme
plt.style.use('seaborn-dark')
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
#to give the title and set colors
plt.title('data of price')
plt.xlabel('time')
plt.ylabel('quantity')
plt.plot(x,y1,color='red',label='Gold')
plt.plot(x,y2,color='green',label='Silver')
#to plot the legend
plt.legend()
plt.show()
```
